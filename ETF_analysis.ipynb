```python
import os
import pandas as pd
import plotly
```

```python
def load_etf_dataframe(tickers):   
    df = pd.DataFrame({'Date' : []})
    for etf in os.listdir('raw'):
        if etf.split('_',)[0] in tickers:
            stock_df = pd.read_csv('raw/'+etf, usecols=['Date', 'Close'])
            stock_df[etf.split('_')[0]] = stock_df['Close']
            stock_df = stock_df.drop(columns='Close')
            df = pd.merge(df, stock_df, on = 'Date', how='outer')
            df = df.sort_values('Date')
            df = df.dropna()

    return df
        
```

```python
etfs = load_etf_dataframe(['XLB', 'XLE', 'XLF', 'XLI', 'XLP', 'XLU', 'XLV', 'XLY', 'SPY'])
etfs.head()
```
